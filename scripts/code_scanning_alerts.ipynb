```python
# Create a table showing the repository name (from `df_repository_filtered['name']`, the total number of alerts, and then additional columns that are the percentage of critical, high, medium, and low alerts

# First add the `repository_name` column to `df_rule_filtered`
df_rule_filtered['repository_name'] = df_repository_filtered['name']

df_repo_alert_statistics = df_rule_filtered.groupby('repository_name').size().to_frame('alert_count')
df_repo_alert_statistics['alert_count'] = df_repo_alert_statistics['alert_count'].astype(float)
df_repo_alert_statistics['percentage_critical'] = df_rule_filtered[df_rule_filtered['security_severity_level'] == 'critical'].groupby('repository_name').size() / df_repo_alert_statistics['alert_count']
df_repo_alert_statistics['percentage_high'] = df_rule_filtered[df_rule_filtered['security_severity_level'] == 'high'].groupby('repository_name').size() / df_repo_alert_statistics['alert_count']
df_repo_alert_statistics['percentage_medium'] = df_rule_filtered[df_rule_filtered['security_severity_level'] == 'medium'].groupby('repository_name').size() / df_repo_alert_statistics['alert_count']
df_repo_alert_statistics['percentage_low'] = df_rule_filtered[df_rule_filtered['security_severity_level'] == 'low'].groupby('repository_name').size() / df_repo_alert_statistics['alert_count']
df_repo_alert_statistics
```

```python
# Create a version of `df_rule_filtered` where the `cwe` list in the `tags` column is exploded into a new column `cwe`, and filter out the rows where `cwe` does not contain the substring `cwe`. The column `tag` currently has data that looks like this: `[correctness, external/cwe/cwe-020, external/cwe/cwe-191, security]`

df_rule_filtered_cwe = df_rule_filtered.explode('tags')
df_rule_filtered_cwe = df_rule_filtered_cwe[df_rule_filtered_cwe['tags'].apply(lambda x: 'cwe' in x)]
df_rule_filtered_cwe['cwe'] = df_rule_filtered_cwe['tags'].apply(lambda x: x.split('/')[2])
df_rule_filtered_cwe
```

```python
# getting the CWE categories for hardware, software, and research just to be safe
cwe1 = pd.read_csv('https://cwe.mitre.org/data/csv/699.csv.zip', compression='zip', index_col=False)
cwe2 = pd.read_csv('https://cwe.mitre.org/data/csv/1000.csv.zip', compression='zip', index_col=False)
cwe3 = pd.read_csv('https://cwe.mitre.org/data/csv/1194.csv.zip', compression='zip', index_col=False)

cwe = pd.concat([cwe1, cwe2, cwe3])

cwe = cwe[['Name', 'CWE-ID']]
cwe['cwe_id'] = cwe['CWE-ID'].apply(lambda x: f"cwe-{str(x).zfill(3)}")
cwe_lookup = cwe[['cwe_id', 'Name']].drop_duplicates()
cwe_lookup
```

```python
# join the `df_rule_filtered_cwe` with the `cwe` column name, with the `cwe_lookup` to get the CWE name, rename the `Name` column to `cwe_name`. Preserve the index of `df_rule_filtered_cwe` 

df_rule_filtered_cwe_with_name = df_rule_filtered_cwe.join(cwe_lookup.rename(columns={'Name': 'cwe_name'}, copy=True).set_index('cwe_id'), on='cwe')

df_rule_filtered_cwe_with_name
```

```python
df_rule_filtered_cwe_with_name.groupby(['cwe_name', 'cwe', 'security_severity_level']).size().unstack().sort_values('critical', ascending=False).head(10)
```
